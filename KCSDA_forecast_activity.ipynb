#### Выбор метрики

Активность конкретного пользователя напрямую зависит от времени, проведённом в сервисе. Время можно связать с количеством действий пользователя

Посмотрим как меняется активность среднего пользователя в ленте с течением времени

```python
# Подневные данные
# (без 0го и текущего дня - в силу нерепрезентативности выборок)

query = f'''
WITH 
    users_activity_t AS (
        SELECT
            toDate(time) AS date,
            user_id,
            countIf(action = 'like') AS user_likes,
            countIf(action = 'view') AS user_views
        FROM {db}.feed_actions
        GROUP BY date, user_id
    ),
    date_stats_t AS (
        SELECT
            date,
            COUNT(DISTINCT user_id) AS uniq_users,
            SUM(user_likes) AS likes,
            SUM(user_views) AS views,
            likes/views AS ctr,
            AVG(user_likes) AS likes_mean,
            AVG(user_views) AS views_mean
        FROM users_activity_t
        GROUP BY date
        ORDER BY date
    ),
    min_date_t AS (
        SELECT MIN(date) AS date
        FROM date_stats_t
    ),
    date_posts_t AS (
        SELECT
            toDate(time) AS date,
            COUNT(DISTINCT post_id) AS uniq_posts
        FROM {db}.feed_actions
        GROUP BY date
    ),
    users_hour_activity_t_1 AS (
        SELECT
            toDate(time) AS date,
            toStartOfHour(time) AS hour,
            COUNT(DISTINCT user_id) AS hour_uniq_users
        FROM {db}.feed_actions
        GROUP BY date, hour
    ),
    users_hour_activity_t AS (
        SELECT
            date,
            MAX(hour_uniq_users) AS hour_max_uniq_users
        FROM users_hour_activity_t_1
        GROUP BY date
    )

SELECT
    l.date AS date,
    l.uniq_users AS uniq_users,
    l.likes AS likes,
    l.views AS views,
    l.ctr AS ctr,
    l.likes_mean AS likes_mean,
    l.views_mean AS views_mean,
    r.uniq_posts AS uniq_posts,
    r2.hour_max_uniq_users AS hour_max_uniq_users
FROM date_stats_t AS l
    LEFT JOIN date_posts_t AS r
        ON l.date=r.date
    LEFT JOIN users_hour_activity_t AS r2
        ON l.date=r2.date
WHERE 
    l.date > (SELECT * FROM min_date_t)
    AND l.date < toDate(now())
'''

df = ph.read_clickhouse(query=query, connection=connection)
df['uniq_users_ratio'] = (df.hour_max_uniq_users / df.uniq_users).mul(100).round(1)  # Отношение максимального количества почасовых уникальных пользователей к числу уникальных дневных пользователей в %
df.head()
```

#### Бэктестинг

```python
from orbit.diagnostics.backtest import BackTester #основной класс для бэктестинга 
from orbit.utils.params_tuning import grid_search_orbit #для подбора оптимальных параметров
```

```python
def get_test_train_len(
    df_len, 
    n,
    test_proportion=0.3
):
    test_len = int(df_len / (1/test_proportion + n - 1))
    train_len = df_len - n * test_len
    
    return test_len, train_len
```

```python
# Expanding
n_max_trainings = 10  # Максимальное количество тренировок

rmsse_vals = []
train_lens = []
test_lens = []

n_trainings = range(1, n_max_trainings+1)
for n in n_trainings:
    incremental_len, min_train_len = get_test_train_len(train_len, n)
    forecast_len = incremental_len

    bt = BackTester(
        model=map_model, # Модель
        df=train_df, # df для обучения
        min_train_len=min_train_len, # Самый маленький кусочек данных, с которого начинается обучение
        incremental_len=incremental_len, # На сколько происходит прирост тренировочных данных
        forecast_len=forecast_len, # Горизонт пронозирования
        window_type="expanding" # Тип окошка
    )
    bt.fit_predict() # Обучаем

    m_df = bt.score() # Метрики
    rmsse_val = m_df[m_df.metric_name == 'rmsse'].metric_values.values[0]
    
    rmsse_vals.append(rmsse_val)
    train_lens.append(min_train_len)
    test_lens.append(forecast_len)
```

```python
sns.lineplot(y=rmsse_vals, x=n_trainings);
plt.title('rmsse');
plt.xlabel('n_sets');
plt.ylabel('rmsse');
```
